```python
#Imports
import gensim #word2vec modeling
import numpy as np #linear algebra 
import pandas as pd #data processing and visualization
import matplotlib.pyplot as plt #graphing
import matplotlib.gridspec as gridspec #graphin
from sklearn.preprocessing import LabelEncoder #convert categorical labels
from sklearn.linear_model import LogisticRegression #logistic regression classification model

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
#import training data to that will be used to train a genism word2vec model, and test data to test its performance.
data = pd.read_json('./train.json')
test = pd.read_json('./test.json')
```

```python
#The shape of the data frames. The train data has 3 columns and ~40k rows.
#The test data has ~10k rows with 2 columns.
print('Training data shape: {}'.format(data.shape))
print('Test data shape: {}'.format(test.shape))
```

```python
# Target variable: the correct cusine label for a recipe's list of ingredients. 
# y-values used to train logistic regression model
target = data.cuisine
```

```python
#Creates an add'l column that stores the number of counts for each row of recipe's ingredients.
data['ingredient_count'] = data.ingredients.apply(lambda x: len(x))
```

```python
# Flattens the nested list into a singular list
#[[1.0], [0.7071067811865475], [0.7071067811865475]] => [1.0, 0.7071067811865475, 0.7071067811865475]
def flatten_lists(lst):
    """Remove nested lists."""
    return [item for sublist in lst for item in sublist]
```

```python
#Data Analysis: Visual Representation of ingredients
f = plt.figure(figsize=(14,8))
gs = gridspec.GridSpec(2, 2)

#Freq of ingredient counts 
ax1 = plt.subplot(gs[0, :])
data.ingredient_count.value_counts().hist(ax=ax1)
ax1.set_title('Recipe richness', fontsize=12)

#The ingredients most common in recipes; Salt would be a poor indictor of a recipe's cuisine.
ax2 = plt.subplot(gs[1, 0])
pd.Series(flatten_lists(list(data['ingredients']))).value_counts()[:20].plot(kind='barh', ax=ax2)
ax2.set_title('Most popular ingredients', fontsize=12)

#Average number of ingredients in cuisines. Low std. deviation
ax3 = plt.subplot(gs[1, 1])
data.groupby('cuisine')['ingredient_count'].mean().sort_values(ascending=False).plot(kind='barh', ax=ax3)
ax3.set_title('Average number of ingredients in cuisines', fontsize=12)

plt.show()
```

```python
# Feed a word2vec with the ingredients
# Trains a word2vec model using the list of ingredients. The vectors have a dimensionality of 350.
# window: The maximum distance between a target word and words around the target word. 
# A higher size of 10 captures the word's usage
# within a sentence. Min_count ignores words that are infrequent within the data as it is harder to capture context
# with less data. Epoch = iterations over corpus.
w2v = gensim.models.Word2Vec(list(data.ingredients), vector_size=350, window=10, min_count=2, epochs=20)
```

```python
#Returns a list of the words with the highest cosine similarity to 'meat'.
#The words have similar embeddings to a meat vector. Contain similar context.
#A higher value means there is a greater similarity.
w2v.wv.most_similar(['meat'])
```

```python
#Returns a list of the words with the highest cosine similarity to 'chicken'.
#The words have similar embeddings to a chicken vector; Contain similar context.
#A higher value means there is a greater similarity.
w2v.wv.most_similar(['chicken'])
```

```python
#Create a word embedding that captures the representation of each document by averaging the list of individual word
#vectors.
def document_vector(doc):
    """Create document vectors by averaging word vectors. Remove out-of-vocabulary words."""
    doc = [word for word in doc if word in w2v.wv.key_to_index]
    vectors = [w2v.wv[word] for word in doc]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros_like(w2v.vector_size) 
```

```python
#Creates a new column that stores each row's averaged word vector.
data['doc_vector'] = data.ingredients.apply(document_vector)
test['doc_vector'] = test.ingredients.apply(document_vector)
```

```python
#Maps the string cuisine labels into unique numbers; class labels
lb = LabelEncoder()
y = lb.fit_transform(target)
```

```python
#X: the data vectors that will be used to train the machine learning model.
X = list(data['doc_vector'])
#x_test: Represents the vectors that the machine learning model will use to evaluate its performance.
X_test = list(test['doc_vector'])
```

```python
# Create a logistic regression classifier
clf = LogisticRegression(C=100)
```

```python
#Trains the logistic regression model with the training data
#X: The data that the model will use to learn patterns and make predictions.
#y: The expected labels that the model will try to predict.
clf.fit(X, y)
```

```python
#Now that the model has been given training data, it can be used to make informed predictions on the classification
#of unseen data which it is tested upon.
#The model makes predictions on the cuisine based on the averaged vector word vector. Cuisines are likely
# to re-use the similar ingredients so the model will classify the test recipe to whichever cuisine has the highest
# match.
y_test = clf.predict(X_test)
y_pred = lb.inverse_transform(y_test)
```

```python
#Outputs a list of recipe predictions made by the model
test_id = [id_ for id_ in test.id]
sub = pd.DataFrame({'id': test_id, 'cuisine': y_pred}, columns=['id', 'cuisine'])
sub.to_csv('clf_output.csv', index=False)
```
